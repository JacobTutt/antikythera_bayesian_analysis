```python
# Import necessary libraries
import jax.numpy as jnp
import pandas as pd
from calender_analysis import Calender_Analysis 
# ------------------ Data ------------------
original_data_path = "data/Hole_Measurements.csv"  
```

## Extra Tests: Investigating the filtering effect

- In this notebook we aim to use the same analysis pipeline that has been implemented in the previous notebooks, however we will compare different data filtering techniques **(None, Basic, Full)**

- In this we use the values found from the optimisations done in the 

## Filtering on Class Initialisation

### Dataset Filtering

The input dataset can be optionally filtered to remove unreliable sections.

This can be done to reduce the dimensionality of the model, without loosing information (data that constrains) on a given paramter. 

**All previous analysis in this pipeline used the Basic filter, however will will test the success of others**

| **Filtering Mode** | **Description** | **No. Sections** | **No. Holes** |
|--------------------|----------------|------------------|--------------|
| **None**  | No filtering applied. | 8 | 81 |
| **Basic** | Removes sections with only one hole. | 6 | 79 |
| **Full**  | Removes sections with fewer than three holes and discards the first and last hole in each section. | 4 | 66 |

```python
nuts_optimal_params_aniso = pd.read_csv('stored_results/Anistropic_Basic_HMC_Opt_Results.csv')
```

# **Anisotropic Models**
## No Filtering

```python
Antikythera_Anisotropic_No_Filter = Calender_Analysis(data = original_data_path, model_type = "anisotropic", filtering = 'None', priors = None, num_cores = 4)
_, _ = Antikythera_Anisotropic_No_Filter.run_hmc_optimised(nuts_optimal_params_aniso, burnin_period=2000, n_samples=12000, n_chains =  4, random_seed=0, save_path = 'stored_results/Anisotropic_None_Opt_NUTS.nc', traceplot=None)
Antikythera_Anisotropic_No_Filter.thinned_hcm_analysis(posterior_data = 'stored_results/Anisotropic_None_Opt_NUTS.nc', summary_table=True, corner_plot=["r", "N", "sigma_r", "sigma_t"])
Antikythera_Anisotropic_No_Filter.plot_credible_intervals(posterior_data = 'stored_results/Anisotropic_None_Opt_NUTS.nc', param="N", percentiles_range=(50, 100))
```

## Full Filtering

```python
Antikythera_Anisotropic_Full_Filter = Calender_Analysis(data = original_data_path, model_type = "anisotropic", filtering = 'Full', priors = None, num_cores = 4)
_, _ = Antikythera_Anisotropic_Full_Filter.run_hmc_optimised(nuts_optimal_params_aniso, burnin_period=2000, n_samples=12000, n_chains =  4, random_seed=0, save_path = 'stored_results/Anisotropic_Full_Opt_NUTS.nc', traceplot=None)
Antikythera_Anisotropic_Full_Filter.thinned_hcm_analysis(posterior_data = 'stored_results/Anisotropic_Full_Opt_NUTS.nc', summary_table=True, corner_plot=["r", "N", "sigma_r", "sigma_t"])
Antikythera_Anisotropic_Full_Filter.plot_credible_intervals(posterior_data = 'stored_results/Anisotropic_Full_Opt_NUTS.nc', param="N", percentiles_range=(50, 100))
```
